```python
plt.figure(figsize=(15, 15))
plt.xticks(rotation=90)
plt.bar(df.loc[:72,'ULB Name'],df.loc[:72,'Public Toilets'],label='Public Toilets')
plt.bar(df.loc[:72,'ULB Name'],df.loc[:72,'Community Toilets'],bottom=df.loc[:72,'Public Toilets'],label='Community Toilets')
plt.bar(df.loc[:72,'ULB Name'],df.loc[:72,'She Toilets'],bottom=df.loc[:72,'Public Toilets']+df.loc[:72,'Community Toilets'],label='She Toilets')
plt.xlabel('ULB Name')
plt.ylabel("Toilets")
plt.title("Toilets in Telangana")

plt.legend()
```

```python
plt.figure(figsize=(15, 20))
plt.xticks(rotation=90)
plt.bar(df.loc[:72,'ULB Name'],df.loc[:72,'ODF Declared Wards'],label='ODF Declared Wards')
plt.bar(df.loc[:72,'ULB Name'],df.loc[:72,'ODF Non Declared Wards'],bottom=df.loc[:72,'ODF Declared Wards'],label='ODF Non Declared Wards')

plt.xlabel('ULB Name')
plt.ylabel("ODF Declared Wards")
plt.title("ODF wards in Telangana")
plt.legend()
```

```python
plt.figure(figsize=(15, 20))
plt.xticks(rotation=90)
plt.bar(df.loc[:72,'ULB Name'],df.loc[:72,'% of Door to Door Collection'],label='% of Door to Door Collection')


plt.xlabel('ULB Name')
plt.ylabel("Waste Collection")
plt.title("waste collection in Telangana")
plt.legend()
```
